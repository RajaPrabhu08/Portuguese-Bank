# sampling

```python
## calcualting the count of yes and no in the y column
```

```python
yes_count = (data['y'] == 'yes').sum()
no_count = (data['y'] == 'no').sum()
```

```python
total_count = len(data)
```

```python
percentage_yes = (yes_count / total_count) * 100
percentage_no = (no_count / total_count) * 100
print("before sampling")
print("Percentage of 'yes' in 'y':", percentage_yes)
print("Percentage of 'no' in 'y':", percentage_no)
```

```python
from imblearn.over_sampling import SMOTE

smote = SMOTE(random_state=123)
X_resampled, y_resampled = smote.fit_resample(x, y)
```

```python
# Concatenate the resampled 'X' and 'y' to form a new DataFrame
```

```python
resampled_data = pd.concat([X_resampled, pd.Series(y_resampled, name='y')], axis=1)
```

```python
# Calculate the count of 'yes' and 'no' in the 'y' column
```

```python
yes_count = (resampled_data['y'] == 1).sum()
no_count = (resampled_data['y'] == 0).sum()
```

```python
# Calculate the total count of values in the 'y' column
```

```python
total_count = len(resampled_data)
```

```python
# Calculate the percentage of 'yes' and 'no'
```

```python
percentage_yes = (yes_count / total_count) * 100
percentage_no = (no_count / total_count) * 100
print("After sampling")
print("Percentage of 'yes' in 'y':", percentage_yes)
print("Percentage of 'no' in 'y':", percentage_no)
```

# sampling resampled data

```python
# Separate features (X) and the target variable (y) from the resampled data
```

```python
X_resampled = resampled_data.drop('y', axis=1)
y_resampled = resampled_data['y']
```

## Summary

1.Logistic Regression delivers an overall test accuracy of 87% with balanced precision, recall, and F1‐score values at around 0.87. While it is fast and interpretable, it appears to be less capable of capturing the non‐linear relationships in your data. 

2.The Decision Tree Classifier (even after parameter tuning) shows an improved test set performance of 93% accuracy with corresponding precision, recall, and F1‐scores near 0.93. Although better than logistic regression, decision trees are known to be prone to overfitting and can be less robust when used alone.

3.In contrast, both the Random Forest Classifier and the XGBoost Classifier achieve a remarkable 95% accuracy on the test set with consistently high metrics (precision around 0.95–0.96, recall around 0.94–0.96, and F1-scores near 0.95). These ensemble methods improve performance by aggregating the predictions of multiple trees, which helps in reducing variance and capturing complex patterns in the data.

# CONCLUSION

For this project, ensemble-based models are the clear winners. The Random Forest and XGBoost classifiers not only provide a significant boost in accuracy compared to logistic regression and single decision trees but they also maintain exceptional balance between precision and recall across classes. This indicates that they are better at generalizing to unseen data and are more robust in capturing non-linearities inherent in your dataset.

Given that both Random Forest and XGBoost yield nearly the same performance, your choice between them might come down to practical considerations. If you value model interpretability and slightly simpler tuning, Random Forests can be quite appealing-especially as they offer straightforward methods for assessing feature importance. On the other hand, if you are comfortable with more intricate hyperparameter tuning for potentially marginal gains or further customization, XGBoost is an excellent candidate.

The XGBoost model used to predict client subscription to term deposits based on marketing campaign data from the Portuguese bank has shown excellent performance:
